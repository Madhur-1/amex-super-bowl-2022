```python
import numpy as np
import pandas as pd
import missingno as msno
```

```python
x_train = pd.read_csv("Data/train_allx.csv")
x_test = pd.read_csv("Data/val_allx.csv")

y_train = pd.read_csv("Data/train_y.csv")
y_test = pd.read_csv("Data/Sample_Output.csv", names = ['ID', 'Target'], header=None)
```

```python
train = pd.concat([x_train, x_test], axis=0)
train.drop('ID', axis=1 , inplace=True)
```

```python
train.reset_index(inplace=True, drop=True)
```

```python
train[['D_36', 'D_44', 'B_41', 'D_130']].isnull().sum()
```

```python
train_transformed = train.groupby(['D_36', 'D_44', 'B_41', 'D_130']).transform(lambda x: x.fillna(x.mean()))
train[train.drop(['D_36', 'D_44', 'B_41', 'D_130'], axis=1).columns] = train_transformed

train_transformed = train.groupby(['D_36', 'D_44', 'B_41',]).transform(lambda x: x.fillna(x.mean()))
train[train.drop(['D_36', 'D_44', 'B_41'], axis=1).columns] = train_transformed

train_transformed = train.groupby(['D_36', 'D_44', 'D_130',]).transform(lambda x: x.fillna(x.mean()))
train[train.drop(['D_36', 'D_44', 'D_130'], axis=1).columns] = train_transformed

train_transformed = train.groupby(['D_36', 'D_44']).transform(lambda x: x.fillna(x.mean()))
train[train.drop(['D_36', 'D_44',], axis=1).columns] = train_transformed

train_transformed = train.groupby(['D_36']).transform(lambda x: x.fillna(x.mean()))
train[train.drop(['D_36', ], axis=1).columns] = train_transformed
```
