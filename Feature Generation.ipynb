```python
import numpy as np
import pandas as pd
```

```python
x_train = pd.read_feather("Data/x_train__IterativeImputer_10itrs.ftr")
x_test = pd.read_feather("Data/x_test__IterativeImputer_10itrs.ftr")

y_train = pd.read_csv("Data/train_y.csv")
y_test = pd.read_csv("Data/Sample_Output.csv", names = ['ID', 'Target'], header=None)
```

```python
# orig_train = pd.read_csv("Data/train_allx.csv")
# orig_train.drop("ID", axis=1, inplace=True)
orig_test = pd.read_csv("Data/val_allx.csv")
orig_test.drop("ID", axis=1, inplace=True)
```

```python
orig_train.dtypes[orig_train.dtypes!='float']
```

```python
cols_not_floats = ['D_36', 'D_44', 'B_41', 'D_130']
```

```python
x_train
```

```python
colstartswith_dict = {}
for col in orig_train.drop(['D_36', 'D_44', 'B_41', 'D_130'], axis=1).columns:
    if col[0] not in colstartswith_dict:
        colstartswith_dict[col[0]] = [col]
    else:
        colstartswith_dict[col[0]].append(col)
```

```python
colstartswith_dict
```

```python
x_train[colstartswith_dict['B']].sum(axis=1)
```

```python
for col_start, cols in colstartswith_dict.items():
    x_train[f"{col_start}_cols_sum"] = x_train[cols].sum(axis=1)
    x_test[f"{col_start}_cols_sum"] = x_test[cols].sum(axis=1)
```

```python
x_train
```

# Using the Variable Dictionary

```python
variable_dict = pd.read_excel("Data/Variable_Dictionary_Data Science Challenge.xlsx").iloc[2:].reset_index(drop=True)
```

```python
variable_dict['Desc_first_word'] = list(map(lambda x:x.split()[0], variable_dict['Description']))
```

```python
variable_dict.set_index("Variable Name", inplace=True)
```

```python
variable_dict.drop(cols_not_floats, inplace=True)
```

```python
variable_dict
```

```python
variable_groups = variable_dict.groupby("Desc_first_word").groups
```

```python
for col_start, cols in variable_groups.items():
    x_train[f"{col_start}_group"] = x_train[cols].sum(axis=1)
    x_test[f"{col_start}_group"] = x_test[cols].sum(axis=1)
```

```python
x_train
```

```python
x_train.to_feather('Data/x_train__IterativeImputer_genFeats.ftr')
y_train.to_feather('Data/x_test__IterativeImputer_genFeats.ftr')
```

# Adding Null Identifiers

```python
x_train.columns.values
```

```python
orig_train.columns.values
```

```python
trainNa = orig_train.drop(['D_36', 'D_44'], axis=1).isna()
trainNa.columns = [f"{c}_missing" for c in trainNa.columns]

train = pd.concat([x_train, trainNa], axis=1)

testNa = orig_test.drop(['D_36', 'D_44'], axis=1).isna()
testNa.columns = [f"{c}_missing" for c in testNa.columns]

test = pd.concat([x_test, testNa], axis=1)
```

```python
train
```

```python
train.to_feather('Data/x_train__IterativeImputer_genFeats.ftr')
test.to_feather('Data/x_test__IterativeImputer_genFeats.ftr')
```
